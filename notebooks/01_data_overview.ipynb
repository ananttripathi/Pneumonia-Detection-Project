# Pneumonia Detection Project - Data Overview

## Business Context

Pneumonia is one of the leading causes of morbidity and mortality worldwide. This project aims to develop an AI-driven system to detect pneumonia from chest X-ray images, assisting healthcare professionals in making accurate and timely diagnoses.

```python
# Import necessary libraries
import sys
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Add src to path
sys.path.append(os.path.join(os.path.dirname(os.getcwd()), 'src'))

from src.data.data_loader import load_labels_from_csv, create_label_mapping
from src.utils.config import BASE_DIR

# Set style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

## 1. Load and Explore Data Files

```python
# Load CSV files
# Note: Update paths to point to pneumonia_files folder on Desktop
pneumonia_files_path = os.path.join(os.path.expanduser('~'), 'Desktop', 'pneumonia_files')

train_labels_path = os.path.join(pneumonia_files_path, 'stage_2_train_labels.csv')
class_info_path = os.path.join(pneumonia_files_path, 'stage_2_detailed_class_info.csv')

# Load data
train_labels_df = load_labels_from_csv(train_labels_path)
class_info_df = load_labels_from_csv(class_info_path)

print("Train Labels DataFrame Shape:", train_labels_df.shape)
print("\nTrain Labels DataFrame Info:")
print(train_labels_df.info())
print("\nFirst few rows:")
print(train_labels_df.head())
```

```python
print("Class Info DataFrame Shape:", class_info_df.shape)
print("\nClass Info DataFrame Info:")
print(class_info_df.info())
print("\nFirst few rows:")
print(class_info_df.head())
print("\nUnique classes:")
print(class_info_df['class'].unique())
print("\nClass distribution:")
print(class_info_df['class'].value_counts())
```

## 2. Check Data Shape and Structure

```python
# Merge dataframes to get complete information
merged_df = train_labels_df.merge(class_info_df, on='patientId', how='left')

print("Merged DataFrame Shape:", merged_df.shape)
print("\nMerged DataFrame Columns:", merged_df.columns.tolist())
print("\nMissing values:")
print(merged_df.isnull().sum())
print("\nData types:")
print(merged_df.dtypes)
```

## 3. Summary Statistics

```python
# Get unique patient IDs
unique_patients = merged_df['patientId'].nunique()
print(f"Total unique patients: {unique_patients}")
print(f"Total records: {len(merged_df)}")

# Class distribution
print("\nClass Distribution:")
class_dist = merged_df['class'].value_counts()
print(class_dist)
print("\nClass Distribution (Percentage):")
print((class_dist / len(merged_df) * 100).round(2))
```

```python
# Visualize class distribution
plt.figure(figsize=(10, 6))
class_dist.plot(kind='bar', color=['steelblue', 'coral', 'lightgreen'])
plt.title('Class Distribution in Dataset', fontsize=14, fontweight='bold')
plt.xlabel('Class', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)

# Add percentage labels
for i, (idx, val) in enumerate(class_dist.items()):
    plt.text(i, val + max(class_dist) * 0.01, 
             f'{val}\n({val/len(merged_df)*100:.1f}%)',
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```
